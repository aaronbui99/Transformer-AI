### Transformer encoder structure



There are two main components in the Encoder Transformer architecture:
1. Self-Attention
2. Feed Forward Network (FFN)

The self-attention mechanism allows each input token to attend to all other tokens within its context, enabling it to capture dependencies and relationships between them. This is achieved by computing attention weights based on the similarity of the query, key, and value vectors for each pair of tokens. The attention weights determine how much information from one token should be considered when processing another. For example, if we have an input sequence "Hello world", the self-attention mechanism would allow each word to attend to all other words in the sentence such as "world" attending to "hello" and compute their importance such as "hello" being more important than "world". This can help the model understand the meaning of each word in relation to others. 

The feed-forward network consists of a series of fully connected layers that transform the output of the self-attention mechanism into a higher-dimensional space such as for "Hello world" the FFN might convert it to something like [0.56789, 0.34567] which represents some kind of feature vector or embedding. These layers help capture complex patterns and interactions within the dat such as "Hello world" having a certain pattern or interaction with other words such as "goodbye world".

Positional encoding is a technique used in transformer models to provide information about the position of tokens within a sequence. It helps the model understand the relative positions of words or subwords, which can be crucial for tasks like language translation and text summarization. For example, consider the sentence "The cat sat on the mat." The positional encoding would assign different values to each word based on its position in the sentence. This allows the model to capture the sequential nature of the input data and make more accurate predictions.

Note: Understand that the transformer model allows parallel processing of all tokens in a sequence, unlike RNNs where computations are done sequentially where each token depends on the previous one. Therefore, it's important to incorporate some form of positional encoding into the model architecture.

*** Self-Attention ***

Self-attention is a mechanism that allows the model to weigh the importance of different parts of an input sequence when generating output. It enables the model to focus on relevant information and ignore irrelevant details, improving its ability to understand complex relationships within data.

For example, "I am a robot" is a sentence with two words: "I" and "robot". The self-attention mechanism can help the model determine which word is more important in understanding the meaning of the sentence. In this case, it might assign higher attention weights to the word "robot", as it carries more semantic significance compared to "I". This helps the model generate accurate predictions or representations based on the context provided by both words.

Let's take a look at meaning of each word in the sentence in self-attention:

| Word     | Attention Weight |
| -------- | ---------------  |
| "I":     |      0.2         |
| "am":    |      0.3         |
| "a":     |      0.1         |
| "robot": |      0.4         |  

In this example, the self-attention mechanism assigns higher attention weights to the word "robot," indicating its greater importance in understanding the overall meaning of the sentence. Second most important word is "am" then word "I" and least important word is "a."
